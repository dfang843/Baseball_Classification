```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batting = pd.read_csv('batting.csv')
fielding = pd.read_csv('fielding.csv')
pitching = pd.read_csv('pitching.csv')
hall_of_fame = pd.read_csv('hall_of_fame.csv')
player_data = pd.read_csv('player.csv')
```

```python
batting.loc[batting['player_id'] == 'fiskca01']
```

```python
batting = batting[batting['ab'].notna()]
```

```python
batting.head()
```

```python
batting = batting.sort_values(['player_id', 'year'])
batting['start_year'] = batting.groupby('player_id')['year'].transform('first')
mask = batting['start_year'] != batting['start_year'].shift()
batting.loc[~mask, 'start_year'] = 0
batting.loc[batting['player_id'] == 'winfida01']
```

```python
pitching = pitching.sort_values(['player_id', 'year'])
pitching['start_year'] = pitching.groupby('player_id')['year'].transform('first')
mask = pitching['start_year'] != pitching['start_year'].shift()
pitching.loc[~mask, 'start_year'] = 0
```

```python
fielding = fielding.sort_values(['player_id', 'year'])
fielding['start_year'] = fielding.groupby('player_id')['year'].transform('first')
mask = fielding['start_year'] != fielding['start_year'].shift()
fielding.loc[~mask, 'start_year'] = 0
```

```python
hall_of_fame = hall_of_fame.loc[(hall_of_fame['inducted'] == 'Y') & (hall_of_fame['category'] == 'Player')]
hall_of_fame.head()
```

```python
career_batting = batting.groupby('player_id').sum()
career_batting = career_batting.drop(columns = ['year', 'team_id', 'league_id', 'stint'], axis = 1)
career_batting = pd.merge(career_batting, player_data[['player_id', 'name_first', 'name_last']], on = 'player_id', how = 'outer')
career_batting = career_batting.dropna()
career_batting
```

```python
career_batting.loc[career_batting['player_id'] == 'fiskca01']
```

```python
career_pitching = pitching.groupby('player_id').sum()
career_pitching = career_pitching.drop(columns = ['year', 'team_id', 'league_id', 'stint'], axis = 1)
career_pitching['ip'] = career_pitching['ipouts'] // 3
career_pitching['era'] = (9 * career_pitching['er']) / career_pitching['ip']
career_pitching = pd.merge(career_pitching, player_data[['player_id', 'name_first', 'name_last']], on = 'player_id', how = 'outer')
career_pitching = career_pitching.dropna()
career_pitching
```

```python
career_fielding = fielding.groupby('player_id').sum()
career_fielding = career_fielding.drop(columns = ['year', 'team_id', 'league_id', 'stint', 'pos'], axis = 1)
career_fielding = pd.merge(career_fielding, player_data[['player_id', 'name_first', 'name_last']], on = 'player_id', how = 'outer')
career_fielding = career_fielding.dropna()
career_fielding
```

```python
joined_batting = pd.merge(career_batting, hall_of_fame[['player_id', 'yearid', 'inducted']], on = 'player_id', how = 'outer')
joined_fielding = pd.merge(career_fielding, hall_of_fame[['player_id', 'yearid', 'inducted']], on = 'player_id', how = 'outer')
joined_pitching = pd.merge(career_pitching, hall_of_fame[['player_id', 'yearid', 'inducted']], on = 'player_id', how = 'outer')
```

```python
joined_batting
```

```python
joined_batting['inducted'] = joined_batting['inducted'].fillna('N')
joined_fielding['inducted'] = joined_fielding['inducted'].fillna('N')
joined_pitching['inducted'] = joined_pitching['inducted'].fillna('N')
joined_batting
```

```python
temp = joined_batting.loc[joined_batting['inducted'] == 'Y']
temp_train = temp.loc[temp['start_year'] < 1975]
temp_test = temp.loc[temp['start_year'] >= 1975]
temp_nan = temp.loc[temp['g'].isna()]
```
